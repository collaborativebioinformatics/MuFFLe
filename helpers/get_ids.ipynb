```python
import os

def get_ids(root_dir):
    # Returns a list of the folder names which serve as your IDs
    return [d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))]

# Usage
ids = get_ids('../../rna-cd-data/')
print(sorted(ids))
# Output: ['3B_426', '3B_428', '3B_429', '3B_431']
# Extracts the digits after the underscore
numeric_ids = [d.split('_')[-1] for d in ids]
# print(sorted(numeric_ids))
# Output: ['426', '428', '429', '431']
print(len(ids))
```
